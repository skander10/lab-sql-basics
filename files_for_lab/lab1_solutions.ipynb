```python
import sqlite3

conn = sqlite3.connect("lab1_bank.sqlite")
cursor = conn.cursor()
```

```python
# Query 1: Erste 5 client_ids mit district_id = 1
cursor.execute("""
SELECT client_id
FROM client
WHERE district_id = 1
ORDER BY client_id ASC
LIMIT 5;
""")
cursor.fetchall()
```

```python
# Query 2: Letzter client_id mit district_id = 72
cursor.execute("""
SELECT client_id
FROM client
WHERE district_id = 72
ORDER BY client_id DESC
LIMIT 1;
""")
cursor.fetchall()
```

```python
# Query 3: 3 niedrigste amounts in der loan-Tabelle
cursor.execute("""
SELECT amount
FROM loan
ORDER BY amount ASC
LIMIT 3;
""")
cursor.fetchall()
```

```python
# Query 4: Mögliche status-Werte in loan, alphabetisch aufsteigend
cursor.execute("""
SELECT DISTINCT status
FROM loan
ORDER BY status ASC;
""")
cursor.fetchall()
```

```python
# Query 5: loan_id der höchsten payments in der loan-Tabelle
cursor.execute("""
SELECT loan_id
FROM loan
ORDER BY payments DESC
LIMIT 1;
""")
cursor.fetchall()
```

```python
# Query 6: account_id und amount der 5 niedrigsten account_ids in loan
cursor.execute("""
SELECT account_id, amount
FROM loan
ORDER BY account_id ASC
LIMIT 5;
""")
cursor.fetchall()
```

```python
# Query 7: account_ids mit niedrigstem amount bei duration=60 in loan
cursor.execute("""
SELECT account_id
FROM loan
WHERE duration = 60
ORDER BY amount ASC
LIMIT 5;
""")
cursor.fetchall()
```

```python
# Query 8: Unique k_symbol Werte in der order-Tabelle
cursor.execute("""
SELECT DISTINCT k_symbol
FROM "order"
ORDER BY k_symbol ASC;
""")
cursor.fetchall()
```

```python
# Query 9: order_ids für account_id 34 in der order-Tabelle
cursor.execute("""
SELECT order_id
FROM "order"
WHERE account_id = 34
ORDER BY order_id ASC;
""")
cursor.fetchall()
```

```python
# Query 10: account_ids für order_id zwischen 29540 und 29560 (order-Tabelle)
cursor.execute("""
SELECT DISTINCT account_id
FROM "order"
WHERE order_id BETWEEN 29540 AND 29560
ORDER BY account_id ASC;
""")
cursor.fetchall()
```

```python
# Query 11: Beträge (amount) gesendet an account_to 30067122 (order-Tabelle)
cursor.execute("""
SELECT amount
FROM "order"
WHERE account_to = 30067122;
""")
cursor.fetchall()
```

```python
# Query 12: 10 erste Transaktionen von account_id 793, neueste zuerst (trans-Tabelle)
cursor.execute("""
SELECT trans_id, date, type, amount
FROM trans
WHERE account_id = 793
ORDER BY date DESC
LIMIT 10;
""")
cursor.fetchall()
```

```python
# Query 13: Anzahl Clients pro district_id < 10 (client-Tabelle)
cursor.execute("""
SELECT district_id, COUNT(*)
FROM client
WHERE district_id < 10
GROUP BY district_id
ORDER BY district_id ASC;
""")
cursor.fetchall()
```

```python
# Query 14: Anzahl Karten pro type, absteigend nach Häufigkeit (card-Tabelle)
cursor.execute("""
SELECT type, COUNT(*)
FROM card
GROUP BY type
ORDER BY COUNT(*) DESC;
""")
cursor.fetchall()
```

```python
# Query 15: Top 10 account_ids nach Summe der loan amounts (loan-Tabelle)
cursor.execute("""
SELECT account_id, SUM(amount)
FROM loan
GROUP BY account_id
ORDER BY SUM(amount) DESC
LIMIT 10;
""")
cursor.fetchall()
```

```python
# Query 16: Anzahl Kredite pro Tag, vor 930907 (exklusiv), absteigend (loan-Tabelle)
cursor.execute("""
SELECT date, COUNT(*)
FROM loan
WHERE date < 930907
GROUP BY date
ORDER BY date DESC;
""")
cursor.fetchall()
```

```python
# Query 17: Kredite pro Tag + duration im Dezember 1997 (loan-Tabelle)
cursor.execute("""
SELECT date, duration, COUNT(*)
FROM loan
WHERE date BETWEEN 971201 AND 971231
GROUP BY date, duration
ORDER BY date ASC, duration ASC;
""")
cursor.fetchall()
```

```python
# Query 18: Summe amount pro type für account_id 396 (trans-Tabelle)
cursor.execute("""
SELECT account_id, type, SUM(amount) AS total_amount
FROM trans
WHERE account_id = 396
GROUP BY type
ORDER BY type ASC;
""")
cursor.fetchall()
```
